```python
from qiskit import *
from qiskit.quantum_info import Statevector
```

```python
circ_BS1 = QuantumCircuit(2)
circ_BS1.h(0)
circ_BS1.cx(0, 1)
circ_BS1.draw('mpl')
```

```python
state = Statevector.from_instruction(circ_BS1)
state.draw('latex')
```

```python
circ_BS2 = QuantumCircuit(2)
circ_BS2.h(0)
circ_BS2.cx(0, 1)
circ_BS2.z(1)
circ_BS2.draw('mpl')
```

```python
state = Statevector.from_instruction(circ_BS2)
state.draw('latex')
```
