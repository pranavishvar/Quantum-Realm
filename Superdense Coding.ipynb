```python
from qiskit import *
from qiskit_aer import Aer
```

```python
circ = QuantumCircuit(2, 2)
circ.h(0)
circ.cx(0, 1)
circ.barrier()

# 2 classical bits
b0 = 1
b1 = 0

if b0:
  circ.x(0)
if b1:
  circ.z(0)


circ.barrier()
circ.cx(0, 1)
circ.h(0)
circ.measure([0, 1], [0, 1])
circ.draw("mpl")
```

```python
simulator = Aer.get_backend("qasm_simulator")
tc = transpile(circ, simulator)
job = simulator.run(tc, shots =1024)
result = job.result()
print(result.get_counts())
```
