```python
import numpy as np
import math
```

```python
num_of_bits = 20
num_of_strings = int(math.pow(2, num_of_bits))
bit_strings = np.array([bin(i)[2:] for i in range(num_of_strings)])
indices = np.random.choice(range(0, num_of_strings), size=int(num_of_strings/2), replace=False)
```

```python
def oracle_b(x):
  if x in bit_strings[indices]:
    return 0
  else:
    return 1

def oracle_c(x):
  return 0

def evaluate(random_samples, oracle, k):
  result = 0

  for elem in random_samples:
    if oracle == 1:
      result += oracle_b(elem)
    elif oracle == 0:
      result += oracle_c(elem)
    else:
      print("Invalid")
      return None


  if result == 0 or result == 7:
    return "Constant"
  else:
    return "Balanced"
```

```python
k = 2
for x in range(20):
  random_sample_indices = np.random.choice(range(0, num_of_strings), size=k, replace=False)
  random_samples = bit_strings[random_sample_indices]
  result = evaluate(random_samples, 1, k)
  print(f"The Function is {result}.")
```
